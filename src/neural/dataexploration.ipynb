# Imports

```python
from dataloader import DataLoader
from plotter import Plotter
from scipy.stats import pearsonr
import numpy as np
import matplotlib.pyplot as plt
```

# Loading Data

```python
dataloader = DataLoader()

df_X_train, df_y_train, df_X_val, df_y_val, df_X_test, df_y_test = dataloader.load_data()
```

# Loading Plotter

```python
plotter = Plotter()
```

# Helper function for correlation

```python
def correlation(x, y, title, x_axis, y_axis):
    plotter.scatter_plot(x, y, title, x_axis, y_axis)
    cor = pearsonr(x, y, alternative = "two-sided")
    print("Correaltion = {}".format(cor))
```

# Investigate Correlation between salary and attrition

```python
salary_train = np.array(list(df_X_train["salary"]))
attrition_train = np.array(list(df_y_train["attrition"]))
correlation(salary_train, attrition_train, "Salary vs Attrition", x_axis = "Salary", y_axis = "Attrition")
```

# Investigate Correlation between tenure and attrition

```python
tenure_train = np.array(list(df_X_train["tenure"]))
attrition_train = np.array(list(df_y_train["attrition"]))
correlation(tenure_train, attrition_train, "Tenure vs Attrition", x_axis = "Tenure", y_axis = "Attrition")
```

# Investigate Correlation between one-hot-encoded variabel and Attrition

```python
def get_attrition_lists(X, y):
    categories = ["CEO", "MGR", "CMGR", "SNRMGR", "TSC", "CONS", "SNRTSC", "SNRCONS"]
    dict = {}

    for col in categories:
        dict[col] = y[X[col] == 1].tolist()
    
    return dict
    

dict = get_attrition_lists(df_X_train, np.array(list(df_y_train["attrition"])))


keys = list(dict.keys())
values = list(dict.values())

# Create the boxplot
plt.figure(figsize=(12, 6))
plt.boxplot(values, patch_artist=True)


# Customize plot
plt.xticks(ticks = range(1, 9), labels=[f'{keys[i-1]}' for i in range(1, 9)])
plt.xlabel("Groups")
plt.ylabel("Attritions")
plt.title("Side-by-Side Boxplots")
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Show the plot
plt.show()
```

# Correlation between salary and tenure

```python
salary_train = np.array(list(df_X_train["salary"]))
tenure_train = np.array(list(df_X_train["tenure"]))
correlation(salary_train, tenure_train, "Salary vs Tenure", x_axis = "Salary", y_axis = "Tenure")
```
